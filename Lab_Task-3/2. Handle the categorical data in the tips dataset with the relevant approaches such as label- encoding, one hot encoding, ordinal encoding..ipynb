```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('C:/123/SRMAP/Semester 6/Machine Learning Lab/Task 3/tips_missing_values_replaced_by_mean_value.csv')
```

```python
Encoded_day = LabelEncoder().fit_transform(df.day)
print(Encoded_day)
```

```python
df['Label encoded day'] = Encoded_day
```

```python
df.head(20)
```

```python
Encoded_smoker = LabelEncoder().fit_transform(df.smoker)
```

```python
df['Label encoded smoker'] = Encoded_smoker
```

```python
one_hot_encoded_day = pd.get_dummies(df['day'])
```

```python
df_encoded_day = pd.concat([df['day'], one_hot_encoded_day], axis=1)
```
