```python
import matplotlib.pyplot as plt
```

```python
products = ['A4 papers', 'Black Pen', 'Blue Pen', 'Marker']
sales = [2400, 800, 300, 50]
```

```python
plt.bar(products, sales, color='orange')

plt.xlabel('Products Name')
plt.ylabel('Sales (Qantity)')
plt.title('Product Sales Comparison')

plt.show()
```
