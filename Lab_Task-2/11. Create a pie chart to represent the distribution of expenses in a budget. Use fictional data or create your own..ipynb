```python
import matplotlib.pyplot as plt
```

```python
categories = ['Rent', 'Food', 'Transport', 'Entertainment', 'Utilities', 'Taxes']
expenses = [8000, 6000, 1000, 4000, 2000, 500]
```

```python
plt.pie(expenses, labels=categories, autopct='%1.1f%%', colors=['skyblue', 'grey', 'green', 'violet', 'blue', 'red'], startangle=90)

plt.title('Expense Budget Distribution')

plt.show()
```
