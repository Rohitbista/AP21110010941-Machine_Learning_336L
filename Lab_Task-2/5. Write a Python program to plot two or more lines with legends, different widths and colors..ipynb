```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 2, 3, 2, 1]
```

```python
plt.plot(x, y1, color='green', linewidth=2, label='Line 1')

plt.plot(x, y2, color='red', linewidth=3, label='Line 2')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Multiple Lines with Legends')

plt.legend()

plt.show()
```
