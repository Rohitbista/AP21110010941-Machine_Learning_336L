```python
import matplotlib.pyplot as plt
```

```python
file_path = 'InputForNo4.txt'  
X = []
Y = []

with open(file_path, 'r') as file:
    for line in file:
        x, y = line.strip().split(',')
        X.append(int(x))
        Y.append(int(y))

plt.plot(X, Y)  

plt.xlabel('X Axis') 
plt.ylabel('Y Axis')
plt.title('Line plotted from input text file') 

plt.show()
```
