```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randn(100)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='grey', edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Randomly Generated Data')

plt.show()
```
