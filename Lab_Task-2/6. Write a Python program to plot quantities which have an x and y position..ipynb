```python
import matplotlib.pyplot as plt
```

```python
x_positions = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4.5, 6, 8.5]
y_positions = [2, 3, 4, 2, 3.5, 4, 5, 1.5, 2.5, 4, 5, 4.5]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x_positions, y_positions, color='blue', marker='o', label='Positions')

plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Scatter Plot of Quantities with X and Y Positions')
plt.legend()

plt.show()
```
