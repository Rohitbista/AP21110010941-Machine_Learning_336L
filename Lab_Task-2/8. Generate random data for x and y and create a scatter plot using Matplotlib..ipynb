```python
import matplotlib.pyplot as plt
import numpy as np  
```

```python
np.random.seed(42)  
num_points = 10
x_positions = np.random.rand(num_points) * 2  
y_positions = np.random.rand(num_points) * 5
```

```python
plt.scatter(x_positions, y_positions, color='green', marker='.', label='Random Data')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Random Scatter Plot')

plt.legend()

plt.show()
```
