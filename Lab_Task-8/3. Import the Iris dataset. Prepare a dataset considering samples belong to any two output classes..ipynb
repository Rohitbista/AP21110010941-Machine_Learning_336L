```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
```

```python
# Load the Iris dataset
iris = load_iris()
data = iris.data
target = iris.target
```

```python
# Select samples from two output classes
class_1_indices = np.where(target == 0)[0]
class_2_indices = np.where(target == 1)[0]
```

```python
# Combine the samples from the two selected classes
selected_data = np.vstack((data[class_1_indices], data[class_2_indices]))
selected_target = np.hstack((target[class_1_indices], target[class_2_indices]))
```

```python
# Select any two input attributes (features)
attribute1_index = 0  # index of the first attribute (0 to 3)
attribute2_index = 2  # index of the second attribute (0 to 3)
```

```python
# Scatter plot for the selected attributes
plt.figure(figsize=(8, 6))
plt.scatter(selected_data[selected_target == 0, attribute1_index], selected_data[selected_target == 0, attribute2_index], color='blue', label='Class 0')
plt.scatter(selected_data[selected_target == 1, attribute1_index], selected_data[selected_target == 1, attribute2_index], color='red', label='Class 1')
plt.xlabel('Attribute ' + str(attribute1_index + 1))
plt.ylabel('Attribute ' + str(attribute2_index + 1))
plt.title('Scatter Plot of Two Classes')
plt.legend()
plt.grid(True)
plt.show()
```
