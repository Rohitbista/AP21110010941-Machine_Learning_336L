```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
df = pd.read_csv("C:/123/SRMAP/Semester 6/Machine Learning Lab/Lab_Task-5/weatherdata.csv")
```

```python
df.tail()
```

```python
data = pd.get_dummies(df)
```

```python
data.head()
```

```python
# Assuming the last column is the target column (class)
X = data.iloc[:, :-3]  # Features
y = df.iloc[:, -1]   # Target variable
```

```python
X.head()
```

```python
y.head()
```

```python
# Create Decision Tree classifier
clf = DecisionTreeClassifier()
```

```python
# Train Decision Tree Classifier
clf = clf.fit(X, y)
```

```python
# Classify a new sample
new_sample = pd.DataFrame({
    'Outlook_overcast': [0],
    'Outlook_rain': [1],
    'Outlook_sunny': [0],
    'Temperature_cool': [1],
    'Temperature_hot': [0],
    'Temperature_mild': [0],
    'Humidity_high': [0],
    'Humidity_normal': [1],
    'Wind_strong': [1],
    'Wind_weak': [0]
})
```

```python
predicted_class = clf.predict(new_sample)
print("Predicted class for new sample:", predicted_class)
```
